## What is Pandas?

Pandas is an open-source Python library used for data manipulation, cleaning, analysis, and visualization. It provides powerful data structures like **Series** and **DataFrame**.

## Why Pandas?

- Easy data manipulation
- Data cleaning & preprocessing
- Fast data analysis
- Read/Write CSV, Excel, JSON, SQL
- Data Analytics & Machine Learning

## Features

- Series
- DataFrame
- Data Import & Export
- Data Cleaning
- Data Filtering
- Sorting
- GroupBy Operations
- Merge & Join
- Missing Value Handling
- Statistical Analysis

## Common Functions

- `pd.Series()` – Create Series
- `pd.DataFrame()` – Create DataFrame
- `pd.read_csv()` – Read CSV File
- `pd.read_excel()` – Read Excel File
- `df.head()` – First 5 Rows
- `df.tail()` – Last 5 Rows
- `df.info()` – Dataset Information
- `df.describe()` – Statistical Summary
- `df.shape` – Dataset Shape
- `df.columns` – Column Names
- `df.isnull()` – Check Missing Values
- `df.fillna()` – Fill Missing Values
- `df.dropna()` – Remove Missing Values
- `df.groupby()` – Group Data
- `df.sort_values()` – Sort Data
- `df.merge()` – Merge DataFrames

## Applications

- Data Cleaning
- Data Analysis
- Data Visualization
- Business Intelligence
- Machine Learning

````python
## Filling Missing Values with the mean of the column
df=pd.read_csv('layoffs.csv')
df['total_laid_off_fillNA']= df['total_laid_off'].fillna(df['total_laid_off'].mean())
df
````

````python
df.describe()
````

````python
df
````

````python
df.dtypes
````

````python
## Rename the Column Name-----

df=df.rename(columns={'Sales Date':'Layoff Date'})
df.head(5)
df=df.rename(columns={'total_laid_off_fillNA':'TLO'})
df.head(5)
````

````python
## Changing the Data Type of the Column

df['total_laid_off_NEW']=df['TLO'].astype(int)
df.head(5)
````

````python
## Using Lambda Function 

df['Lambda_fn_Value']=df['TLO'].apply(lambda x:x*2)
df.head()
````

````python
 ## Data Aggregating and Grouping
Comapany_Mean=df.groupby('industry')['TLO'].mean()
print(Comapany_Mean)
````

````python
Country_wise_industry=df.groupby(['industry','country'])['TLO'].agg(['mean'])
(Country_wise_industry)
````

````python
## MUltiple Aggregation fn :
AGG_VALUE=df.groupby('industry')['TLO'].agg(['sum','mean','min','max','count']).astype(int)
AGG_VALUE
````

````python
## Merging Data Frame


data1 = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva","Sam"],
    "Age": [24, 27, 22, 32, 29, 21],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Miami","GKP,UP"]
}
data2 = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva","Appu"],
    "Gender": ["Female", "Male", "Male", "Male", "Female","Female"],
    "Salary": [50000, 60000, 55000, 70000, 65000,100000]
}

# Convert Dictionaries to DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(df1)
print()
print(df2)
````

````python
Inner_TABLE=pd.merge(df1,df2,on="Name",how="inner")
Inner_TABLE
````

````python
Outer_TABLE=pd.merge(df1,df2,on="Name",how="outer")
Outer_TABLE
````

````python
Left_Outer_TABLE=pd.merge(df1,df2,on="Name",how="left")
Left_Outer_TABLE
````

````python
Right_Outer_TABLE=pd.merge(df1,df2,on="Name",how="right")
Right_Outer_TABLE
````

#  Pandas Data Reading Functions

Pandas supports reading data from multiple file formats and data sources.

| Function | Description | Returns |
|----------|-------------|---------|
| `pd.read_csv()` | Read CSV (Comma-Separated Values) file | DataFrame |
| `pd.read_excel()` | Read Excel (.xlsx, .xls) file | DataFrame |
| `pd.read_json()` | Read JSON file | DataFrame |
| `pd.read_html()` | Read HTML tables from a webpage | List of DataFrames |
| `pd.read_sql()` | Read data from SQL query/database | DataFrame |
| `pd.read_sql_query()` | Read SQL query result | DataFrame |
| `pd.read_sql_table()` | Read an entire SQL table | DataFrame |
| `pd.read_xml()` | Read XML file | DataFrame |
| `pd.read_parquet()` | Read Parquet file | DataFrame |
| `pd.read_feather()` | Read Feather file | DataFrame |
| `pd.read_pickle()` | Read Pickle file | Object/DataFrame |
| `pd.read_hdf()` | Read HDF5 file | DataFrame |
| `pd.read_orc()` | Read ORC file | DataFrame |
| `pd.read_sas()` | Read SAS dataset | DataFrame |
| `pd.read_spss()` | Read SPSS (.sav) file | DataFrame |
| `pd.read_stata()` | Read Stata (.dta) file | DataFrame |
| `pd.read_fwf()` | Read Fixed Width Format file | DataFrame |
| `pd.read_table()` | Read Tab-separated (.txt/.tsv) file | DataFrame |
